```python
import nltk
nltk.download('punkt')

# NLTK Corpus 말뭉치 데이터셋 다운로드 받기
nltk.download('all')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
raw_text1 = '''
Happiness in this world, when it comes, comes incidentally.
Make it the object of pursuit, and it leads us a wild-goose chase,
and is never attained.
'''

raw_text2 = '''
No Pain No Gain!
'''
```

```python
# wordpunct_tokenize
# 단어 단위 토큰화
result1 = word_tokenize(raw_text1)
result2 = word_tokenize(raw_text2)
print(result1)
print(result2)
```

```python
raw_text = [raw_text1,raw_text2]
```

```python
# 문장 단위 토큰화
sent_result = sent_tokenize(raw_text1)
sent_result,len(sent_result)
```

##### ex) 여러 문장에서 토큰 추출
---


```python
# 문장 단위로 추출
total_token = []
for sent in raw_text:
  
  # 문장 추출
  sentResult = sent_tokenize(sent)

  # 문장에서 추출한 토큰
  print(f' [ Total sentence ] => {sentResult}')
  print('-------------------------')
  for ele in sentResult:
    print(f'     [ Element sentence ] => {ele}')
    print('-------------------------')
    wordResult = word_tokenize(ele)
    print(f'           [ wordResult ] => {wordResult}')
    print('-------------------------')
  
  print('============================================================================================================')
  # 모든 문장의 토큰에 추가
  total_token.append(wordResult)

print(total_token)    
```

```python
from konlpy.tag import Okt

# 형태소 분리 객체
okt = Okt()
```

```python
# 형태소 분리 : okt.morphs
result = okt.morphs('오늘은 월요일입니다.')
print(result)
```

```python
# 형태소 분리 후 태깅(Tagging) => 품사 : okt.pos
result2 = okt.pos('오늘은 월요일입니다.')
print(result2)
```

```python
# 어간 : okt.pos(,stem = True)
result3 = okt.pos('오늘은 월요일입니다.',stem =True)
print(result3)
```

## [3] 텍스트 벡터화
---
- 텍스트 => 수치화
- 희소행렬(OHE) : BOW 방식 => Count기반, TF-IDF 기반
- 밀집벡터: Embedding 방식, Word2Vect


```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
corpus = [raw_text1]
```

```python
ohe = CountVectorizer()
```

```python
ohe.fit(corpus)
```

```python
result = ohe.transform(corpus)
print(result)
```

```python
result = result.toarray()
print(result)
```

```python
# TF-IDF 기반
tfldf = TfidfVectorizer()
```

```python
tf_corpus = tfldf.fit_transform(corpus)
```

```python
tf_corpus = tf_corpus.toarray()
```

```python
print(tf_corpus)
```

[1] sentense,word 분리 

```python
total_sent='Wiki is in Ward is original description: The simplest online database that could possibly work.\
Wiki is a piece of server software that allows users to freely create and edit Web page content using any Web browser. Wiki supports hyperlinks and has a simple text syntax for creating new pages and crosslinks between internal pages on the fly.\
Wiki is unusual among group communication mechanisms in that it allows the organization of contributions to be edited in addition to the content itself.Like many simple concepts, "open editing" has some profound and subtle effects on Wiki usage. Allowing everyday users to create and edit any page in a Web site is exciting in that it encourages democratic use of the Web and promotes content composition by nontechnical users.'
```

```python
sent = sent_tokenize(total_sent)
print(len(sent))
print(sent)
```

```python
word_list=[]
for ele_sent in sent:
  result = []
  result = word_tokenize(ele_sent)

  for word in result:
    if word in en_stopwords:
      pass
    else:
      word_list.append(word)

print(len(word_list))      

```
