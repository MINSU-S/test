from tensorflow.keras.preprocessing import text

```python
from tensorflow.keras.preprocessing import text
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
raw_text = 'Wiki is in Ward is original description: The simplest online database that could possibly work.\
Wiki is a piece of server software that allows users to freely create and edit Web page content using any Web browser. Wiki supports hyperlinks and has a simple text syntax for creating new pages and crosslinks between internal pages on the fly.\
Wiki is unusual among group communication mechanisms in that it allows the organization of contributions to be edited in addition to the content itself.Like many simple concepts, "open editing" has some profound and subtle effects on Wiki usage. Allowing everyday users to create and edit any page in a Web site is exciting in that it encourages democratic use of the Web and promotes content composition by nontechnical users.'
```

```python
# 토큰으로 나누기
tokens  = text.text_to_word_sequence(raw_text)
print(tokens, len(tokens))
```

```python
myToken = text.Tokenizer()
```

```python
# 단어 빈도수가 낮은 순으로 사전 생성
myToken.fit_on_texts(tokens)
print(myToken.word_index)
```

```python
import matplotlib.pyplot as plt
```

```python
list_keys = list(myToken.word_index.keys())
list_values = list(myToken.word_index.values())
```

```python
plt.figure(figsize=(13,14))
plt.barh(list_keys, list_values, color ='grey', alpha = 0.3)
plt.barh(list_keys[-1], list_values[-1], color ='tomato')
plt.show()
```

```python
sentences = ['Wiki is in Ward is original description: The simplest online database that could possibly work',
'Wiki is a piece of server software that allows users to freely create and edit Web page content using any Web browser',
'Wiki supports hyperlinks and has a simple text syntax for creating new pages and crosslinks between internal pages on the fly.',
'Wiki is unusual among group communication mechanisms in that it allows the organization of contributions to be edited in addition to the content itself.',
'Like many simple concepts, "open editing" has some profound and subtle effects on Wiki usage.',
' Allowing everyday users to create and edit any page in a Web site is exciting in that it encourages democratic use of the Web and promotes content composition by nontechnical users.']
```

```python
tokenizer = text.Tokenizer()

# 단어 빈도수가 높은 순으로 낮은 정수 인덱스 부여
tokenizer.fit_on_texts(sentences)
```

```python
# 단어 : 단어 인덱스
print(tokenizer.word_index)
```

```python
# 단어별 개수
print(tokenizer.word_counts)
```

```python
# 문장을 생성된 사전(voca)를 기반으로 수치화
print(tokenizer.texts_to_sequences(sentences))
```

```python
# 문장을 생성된 사전(voca)를 기반으로 수치화
seq_voca = tokenizer.texts_to_sequences(sentences)
print(f'seq_voca: {len(seq_voca)}')
print(seq_voca)
```
